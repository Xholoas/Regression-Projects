```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("advertising.csv")
data.drop(data[["Ad Topic Line","City","Country","Timestamp"]],axis=1,inplace=True)
```

```python
plt.hist(data["Age"],bins=30,alpha=1)
```

```python
sns.jointplot(x=data["Age"],y=data["Area Income"])
```

```python
sns.pairplot(data,hue="Male")
```

```python
X = data[["Daily Time Spent on Site","Age","Daily Internet Usage","Male","Area Income"]]
y = data["Clicked on Ad"]

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test,prediction))
```
